### 3. EDA
Exploratory data analysis

Разведочный анализ данных (англ. exploratory data analysis, *EDA*) — анализ основных свойств данных, нахождение в них общих закономерностей, распределений и аномалий, построение начальных моделей, зачастую с использованием инструментов визуализации.

Понятие введено математиком Джоном Тьюки, который сформулировал цели такого анализа следующим образом:

* максимальное «проникновение» в данные,
* выявление основных структур,
* выбор наиболее важных переменных,
* обнаружение отклонений и аномалий,
* проверка основных гипотез,
* разработка начальных моделей.

*Основные средства разведочного анализа* — изучение вероятностных распределений переменных, построение и анализ корреляционных матриц, факторный анализ, дискриминантный анализ, многомерное шкалирование.

```python
X_train.describe()
```

**Rooms**

```python
# Проверим данные Rooms
X_train['Rooms'].value_counts()
```

```python
# Заполним фичу Rooms 0, 10, 19 средним значением комнат
# Для этого сделаем срез строк, где значение Rooms == 0 или Rooms > 6 и присвоим этому срезу среднее значение с помощью .median()
X_train.loc[(X_train['Rooms'] == 0) | (X_train['Rooms'] >= 6), 'Rooms'] = X_train['Rooms'].median()
X_train['Rooms'].value_counts()
```

**Square**

```python
# Изучим параметр Square. Выведем сортировку по возрастанию
X_train.sort_values(by='Square')
```

```python
# Посмотрим на графике для наглядности данные в соотношении количества комнат в апартаментах и размер апартаментов в кв м
plt.scatter(X_train['Rooms'], X_train['Square'])
```

```python
# Однокомнатные и двухкомнатные апартаменты с площадью выше 400 кв м явно вброс, найдем значение среднее для такого кол-ва комнат
mean_sq_one_room = X_train.loc[(X_train['Square'] < 400) & (X_train['Rooms'] == 1.0), 'Square'].mean()
mean_sq_two_room = X_train.loc[(X_train['Square'] < 400) & (X_train['Rooms'] == 2.0), 'Square'].mean()
mean_sq_one_room, mean_sq_two_room
```

```python
# Заполнили аномальные значения средним в данной выборке в разрезе количества комнат
X_train.loc[(X_train['Square'] > 400) & (X_train['Rooms'] == 1.0), 'Square'] = mean_sq_one_room
X_train.loc[(X_train['Square'] > 400) & (X_train['Rooms'] == 2.0), 'Square'] = mean_sq_two_room
```

```python
# Посмотрим на график еще раз и видим что график распределен более равномерно
plt.scatter(X_train['Rooms'], X_train['Square'])
```

**LifeSquare**

```python
X_train.sort_values(by='LifeSquare', ascending=False)
```

```python
# Посмотрим на параметр LifeSquare. Во-первых здесь есть NaN и здесь есть аномальные значения
plt.scatter(X_train['Square'], X_train['LifeSquare'])
```

```python
# Там где есть аномальная площадь в 7480.592129 вычислим для нее среднюю
mean_sq_three_room = X_train.loc[(X_train['LifeSquare'] < 7000) & (X_train['Rooms'] == 3.0), 'LifeSquare'].mean()
mean_sq_three_room
```

```python
# Присвоим значение аномальной переменной LifeSquare
X_train.loc[(X_train['LifeSquare'] > 7000) & (X_train['Rooms'] == 3.0), 'LifeSquare'] = mean_sq_three_room
```

```python
# Снова построим график
plt.scatter(X_train['Rooms'], X_train['LifeSquare'])
```

```python
# Посмотрим где есть значения NaN в разрезе на кол-во комнат и присвоим соответствующее значение:
def mean_lsq_room(X, max_rooms):
    for room in range(1, max_rooms + 1):
        mean_lsq_room = X.loc[(np.isnan(X['LifeSquare']) == False) & (X['Rooms'] == room), 'LifeSquare'].mean()
        X.loc[(np.isnan(X['LifeSquare']) == True) & (X['Rooms'] == room), 'LifeSquare'] = mean_lsq_room
```

```python
# Присвоим полученные средния значения по столбцу LifeSquare
mean_lsq_room(X_train, 5)
```

**Floor and HouseFloor**

```python
# Изучим параметр Floor и HouseFloor. Выведем сортировку по возрастанию
X_train.sort_values(by='Floor')
```

```python
# Изучим параметр Floor и HouseFloor. Выведем сортировку по возрастанию
X_train.sort_values(by='HouseFloor')
```

```python
# С Floor все ок, но HouseFloor не может быть меньше чем Floor
X_train.info()
```

```python
# Там где этажность дома < этаж апартаментов мы присвоим этажности дома = этаж апартмаентов
X_train['HouseFloor'] = X_train.loc[(X_train['Floor'] > X_train['HouseFloor']), 'Floor']
```

```python
# Выведем сортировку по возрастанию
X_train.sort_values(by='HouseFloor')
```

```python
floor = X_train.loc[(np.isnan(X_train['HouseFloor'])) == True, 'Floor']
```

```python
X_train.loc[(np.isnan(X_train['HouseFloor'])) == True, 'HouseFloor'] = floor
```

```python
# Выведем сортировку по возрастанию
X_train.sort_values(by='HouseFloor')
```

**HouseYear**

```python
# Выведем сортировку по возрастанию
X_train.sort_values(by='HouseYear')
```

```python
# Найдем значения явно аномальные для параметра HouseYear, для таких HouseYear заменим год на рандомный из списка домов с таким же HouseFloor
for floor in X_train.loc[(X_train['HouseYear'] > 2020), 'HouseFloor']:
    rand_house_year = random.choice(X_train.loc[(X_train['HouseFloor'] == floor), 'HouseYear'].tolist())
    X_train.loc[(X_train['HouseYear'] > 2020) & (X_train['HouseFloor'] == floor), 'HouseYear'] = rand_house_year
```

```python
# Выведем сортировку по возрастанию
X_train.sort_values(by='HouseYear')
```

# Создание модели RandomForestRegressor

```python
X_train.head()
```

```python
to_del_list = ["DistrictId", "LifeSquare", "Healthcare_1", "Ecology_2", "Ecology_3", "Shops_2"]
X_train.drop(to_del_list, axis=1, inplace=True)
X_test.drop(to_del_list, axis=1, inplace=True)

model = RandomForestRegressor(n_estimators=1000, max_depth=16, random_state=42, max_features=7)
model.fit(X_train, y_train)

y_pred_final = model.predict(X_test)

preds_final['Price'] = y_pred_final
preds_final.to_csv('predictions.csv', index=False)
```
